```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

url = 'https://raw.githubusercontent.com/Haeeshart/Python-for-Biologists/refs/heads/main/Global%20difference%20(1).csv'
df = pd.read_csv(url)
df.head()
```

```python
df.dropna(inplace = True)
print(df.shape)
```

```python
df = df.set_index('Gene')
```

```python
sns.clustermap(df, metric="euclidean", z_score=0, method="average",cmap="vanimo")
png_path = 'HeatMAp.png'
pdf_path = 'HeatMAp.pdf'
plt.savefig(png_path, dpi=600)
plt.savefig(pdf_path)
plt.show()

# heatmap shows replicates are nicely clustered amongst each strain. Replicates are similar for both WT and deleted hnox but the strains themselves
# are different from each other
```

```python
sns.clustermap(df, metric="Correlation", z_score=0, method="average", cmap="vanimo")
plt.show()
```
